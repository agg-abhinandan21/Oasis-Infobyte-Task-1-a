```python
# Step 1: Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load the uploaded dataset (Skip if already loaded in Colab)
from google.colab import files
uploaded = files.upload()

# Replace 'menu.csv' with your uploaded file's name
df = pd.read_csv('menu.csv')

# Step 3: Preview data
print("📌 First 5 Rows:")
print(df.head())

print("\n🔍 Dataset Info:")
print(df.info())

print("\n📏 Statistical Summary:")
print(df.describe())

# Step 4: Check for missing values
print("\n🧼 Missing Values:")
print(df.isnull().sum())

# Step 5: Data Types and Column Names
print("\n📦 Data Types:")
print(df.dtypes)

print("\n📋 Columns:")
print(df.columns)

# Step 6: Value Counts for Category Column
print("\n🍽️ Value Counts in 'Category':")
print(df['Category'].value_counts())

# Step 7: Correlation Heatmap
plt.figure(figsize=(12,8))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap='coolwarm')
plt.title("📊 Heatmap of Numerical Features")
plt.show()

# Step 8: Boxplots for Fat, Calories, etc.
features = ['Calories', 'Total Fat', 'Saturated Fat', 'Cholesterol', 'Sodium', 'Carbohydrates', 'Sugars', 'Protein']
for feature in features:
    plt.figure(figsize=(10,5))
    sns.boxplot(data=df, x='Category', y=feature)
    plt.title(f'📦 Boxplot of {feature} by Category')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Step 9: Histograms for numerical columns
df[features].hist(figsize=(14,10), bins=15)
plt.suptitle("📉 Histograms of Nutritional Values", fontsize=16)
plt.tight_layout()
plt.show()

# Step 10: Pairplot for Selected Features
sns.pairplot(df[["Calories", "Total Fat", "Carbohydrates", "Sugars", "Protein"]])
plt.suptitle("🔗 Pairplot of Selected Nutrients", y=1.02)
plt.show()

# Step 11: Average Calories by Category
avg_calories = df.groupby('Category')['Calories'].mean().sort_values(ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x=avg_calories.index, y=avg_calories.values)
plt.xticks(rotation=45)
plt.title("🔥 Average Calories per Category")
plt.ylabel("Average Calories")
plt.xlabel("Category")
plt.tight_layout()
plt.show()

# Step 12: Nutritional Value Comparison
df_grouped = df.groupby('Category')[features].mean()
df_grouped.plot(kind='bar', figsize=(15,8))
plt.title("🍱 Average Nutritional Values per Category")
plt.ylabel("Average Value")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("✅ EDA Completed Successfully!")
```
